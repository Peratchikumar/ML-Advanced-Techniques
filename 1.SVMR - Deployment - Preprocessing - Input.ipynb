```python
#Importing Libraries
import pickle #How to save model
import pandas as pd #Pandas is a Library

#Word/Term                    Definition

#import pickle                  →   Loads the pickle library, used to
                                #save or load trained models and
                                #other Python objects to/from disk

#How to save model              →   Inline comment noting that
                                #pickle is used to save models

#import pandas as pd            →   Loads pandas for data manipulation
                                #and DataFrame operations

#Pandas is a Library            →   Inline comment noting that
                                #pandas is a library


#Summary:
#This cell imports the two libraries needed for this section — pickle for saving/loading the model and scaler, and pandas for 
#handling data — setting up the basics before loading a saved model or preprocessing new input.
```

```python
#Loading the Dataset
dataset=pd.read_csv("50_Startups.csv") 

#Word/Term                    Definition

#dataset = pd.read_csv           →   Loads the "50_Startups.csv" file
#("50_Startups.csv")                 into a pandas DataFrame named dataset


#Summary:
#This cell loads the 50 Startups dataset again in this deployment phase — likely needed to demonstrate the full preprocessing 
#pipeline (encoding, splitting) on the same raw data before applying the previously saved scaler and model.
```

```python
#Preprocessing and Splitting the Dataset
dataset=pd.get_dummies(dataset,drop_first=True)
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]
dependent=dataset[["Profit"]]
from sklearn.model_selection import train_test_split  
X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0)

#Word/Term                    Definition

#dataset = pd.get_dummies        →   Converts categorical (text/object)
#(dataset, drop_first=True)          columns — like 'State' — into
                                #numeric dummy/indicator variables,
                                #dropping the first category to
                                #avoid the dummy variable trap

#independent = dataset           →   Selects 5 columns — 'R&D Spend',
#[['R&D Spend',                      'Administration', 'Marketing
#'Administration',                   Spend', 'State_Florida', and
#'Marketing Spend',                  'State_New York' — as the
#'State_Florida',                    independent features (input)
#'State_New York']]

#dependent = dataset              →   Selects the 'Profit' column as
#[["Profit"]]                        the dependent/target variable (output)

#from sklearn.model_selection    →   Imports the function to split
#import train_test_split             data into training and testing subsets

#X_train, X_test,                →   Splits independent and dependent
#y_train, y_test =                   data into training and testing
#train_test_split                    sets, with 30% of the data
#(independent, dependent,            (test_size = 0.30) reserved for
#test_size=0.30,                     testing, and a fixed random_state
#random_state=0)                     for reproducible results


#Summary:
#This cell repeats the full preprocessing pipeline — encoding the 'State' column, splitting into independent/dependent variables, 
#and creating train/test sets — exactly matching the steps used when the original model was trained, ensuring consistency before 
#using the saved scaler and model in this deployment phase.
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)

#Word/Term                    Definition

#from sklearn.preprocessing      →   Imports the StandardScaler class
#import StandardScaler              for standardizing features

#sc = StandardScaler()           →   Creates a NEW StandardScaler object

#X_train = sc.fit_transform      →   Fits this new scaler on the
#(X_train)                           training data and transforms it

#X_test = sc.transform            →   Transforms test data using the
#(X_test)                            same newly-fitted scaler


#Summary:
#This cell creates and fits a brand new StandardScaler on the training data, rather than reusing the one loaded earlier from "scaler.sav".
```

```python
#Scaling New Input Data
preinput=sc.transform([[1300,12000,4000,0,1]])

#Word/Term                    Definition

#preinput = sc.transform         →   Takes new raw input values (R&D
#([[1300,12000,4000,0,1]])           Spend=1300, Administration=12000,
                                #Marketing Spend=4000,
                                #State_Florida=0, State_New York=1)
                                #and scales them using the current sc scaler

#Summary:
#This cell scales a new input row before feeding it into the loaded model for prediction. The input represents a startup 
#based in New York (State_Florida=0, State_New York=1).
```

```python
#Loading the Model and Predicting
loaded_model = pickle.load(open("finalized_model_svm_regression.sav","rb"))
result = loaded_model.predict(preinput)

#Word/Term                    Definition

#loaded_model = pickle.load      →   Loads the previously saved SVR
#(open("finalized_model_             model back from the file, in
#svm_regression.sav", "rb"))         binary read mode ('rb')

#result = loaded_model           →   Uses the reloaded model to
#.predict(preinput)                  predict Profit on the scaled
                                #input (preinput) prepared earlier


#Summary:
#This cell reloads the saved SVR model and uses it to predict Profit on new, already-scaled input data — completing the 
#deployment pipeline in which a model trained once is reused for future predictions without retraining.
```
