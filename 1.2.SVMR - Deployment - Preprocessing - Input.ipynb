```python
#Importing Libraries
import pickle #How to save model
import pandas as pd #Pandas is a Library

#Word/Term                    Definition

#import pickle                  →   Loads the pickle library, used to
                                #save or load trained models and
                                #other Python objects to/from disk

#How to save model              →   Inline comment noting that
                                #pickle is used to save models

#import pandas as pd            →   Loads pandas for data manipulation
                                #and DataFrame operations

#Pandas is a Library            →   Inline comment noting that
                                #pandas is a library


#Summary:
#This cell imports the two libraries needed for this section — pickle for saving/loading the model and scaler, and pandas for 
#handling data — setting up the basics before loading a saved model or preprocessing new input.
```

```python
#Loading the Saved Scaler and Scaling New Input
sc = pickle.load(open("scaler.sav","rb"))
preinput=sc.transform([[1300,12000,4000,0,1]])

#Word/Term                    Definition

#sc = pickle.load                →   Loads the previously saved
#(open("scaler.sav", "rb"))          StandardScaler object back
                                #from the file "scaler.sav", in
                                #binary read mode ('rb')

#preinput = sc.transform          →   Takes new raw input values (R&D
#([[1300,12000,4000,0,1]])           Spend=1300, Administration=12000,
                                #Marketing Spend=4000,
                                #State_Florida=0, State_New York=1)
                                #and scales them using the loaded scaler

#Summary:
#This cell reloads the exact scaler that was fitted on the original training data, then uses it to scale a new input row 
#(a startup based in New York). Since sc is loaded fresh right here — right before preinput is created — this correctly avoids 
#the earlier scaler mix-up issue, ensuring the new input is scaled consistently with how the model was originally trained.
```

```python
#Previewing the Scaled Input
preinput

#Word/Term          Definition

#preinput      →     Displays the scaled version of the new input
                     #values (R&D Spend, Administration, Marketing
                     #Spend, State dummies) after being transformed
                     #by the sc scaler


#Summary:
#This is a quick check of preinput to visually confirm the new input values were scaled before being passed to the loaded model for prediction.
```

```python
#Loading the Model and Predicting
loaded_model = pickle.load(open("finalized_model_svm_regression.sav","rb"))
result = loaded_model.predict(preinput)

#Word/Term                    Definition

#loaded_model = pickle.load      →   Loads the previously saved SVR
#(open("finalized_model_             model back from the file, in
#svm_regression.sav", "rb"))         binary read mode ('rb')

#result = loaded_model           →   Uses the reloaded model to
#.predict(preinput)                  predict Profit on the scaled
                                #input (preinput) prepared earlier


#Summary:
#This cell reloads the saved SVR model and uses it to predict Profit on new, already-scaled input data — completing the 
#deployment pipeline in which a model trained once is reused for future predictions without retraining.
```
