```python
#IMPORTING REQUIRED LIBRARIES
import pickle 
import pandas as pd 

#import pickle          → Loads the pickle library, used to save and load trained models
#import pandas as pd    → Loads the pandas library, used to work with data in table form

#These two libraries let us load a saved model (pickle) and handle data (pandas).
```

```python
#LOADING SAVED SCALERS
sc=pickle.load(open("scaler.sav","rb"))
scy=pickle.load(open("scaler.sav","rb"))

#pickle.load(open("scaler.sav", "rb"))   → Loads a previously saved scaler object from file
#sc = ...                                 → Stores the loaded scaler, used to scale input features
#scy = ...                                → Stores the loaded scaler, used to scale/unscale the target variable

#These scalers were saved during training, so we load them back to keep the same scaling when making new predictions.
```

```python
#LOADING THE SAVED MODEL
loaded_model=pickle.load(open("finalized_model_svm_regression.sav",'rb'))

#pickle.load(open("finalized_model_svm_regression.sav", 'rb'))   → Loads the trained SVM regression model back from file
#loaded_model = ...                                                → Stores the loaded model, ready to make predictions

#This loads the model we already trained and saved, so it can be reused without retraining.
```

```python
#CREATING NEW INPUT DATA FOR PREDICTION
new_data = pd.DataFrame([[1300,14000,4000,0,1]],
        columns= ['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York'])

#pd.DataFrame([[...]], columns=[...])   → Creates a table (DataFrame) with one row of new input values
#[1300, 14000, 4000, 0, 1]               → The actual values for R&D Spend, Administration, Marketing Spend, and State (encoded)
#columns = [...]                          → Names each column so the data matches the format the model was trained on
#new_data = ...                           → Stores this new row of data, ready to be scaled and predicted

#This creates a new sample of data, in the same structure as the training data, so the model can make a prediction on it.
```

```python
#SCALING THE NEW INPUT DATA
preinput = sc.transform(new_data)

#sc.transform(new_data)   → Scales new_data using the same scaler that was used during training
#preinput = ...             → Stores the scaled version of new_data, ready to feed into the model

#The model was trained on scaled data, so new input must also be scaled the same way before predicting.
```

```python
#MAKING A PREDICTION
result = loaded_model.predict(preinput)

#loaded_model.predict(preinput)   → Uses the loaded model to predict an output based on the scaled input data
#result = ...                      → Stores the predicted (scaled) value

#This gives us the model's prediction, but it's still in scaled form — it needs to be converted back to the real value next.
```

```python
#DISPLAYING THE PREDICTED RESULT
result

#result   → Displays the predicted value, still in scaled form

#This shows the raw prediction before converting it back to the original scale.
```

```python
#CONVERTING PREDICTION BACK TO ORIGINAL SCALE
preoutput=scy.inverse_transform(new_data)

#scy                        → The saved scaler object used earlier for the target variable
#.inverse_transform(new_data)   → Converts new_data back to its original (real-world) scale
#preoutput = ...              → Stores the converted value

#This is meant to convert the scaled prediction back to a real value — but note, this line uses new_data instead of result, 
#so double-check if that's intended, since earlier this step was applied on the prediction (result), not the input.
```

```python
#MAKING A PREDICTION USING preoutput
result = loaded_model.predict(preoutput)

#loaded_model.predict(preoutput)   → Uses the loaded model to predict an output, this time using preoutput as input
#result = ...                       → Stores this new predicted value, overwriting the earlier result

#Note: this reuses the result and feeds in preoutput (which was meant to be the final unscaled output) 
#back into the model — this looks like a possible mix-up in the code flow, worth double-checking against your intended steps.
```

```python
#DISPLAYING THE RESULT
result

#result   → Displays the value currently stored in result

#Just shows whatever was last stored in result from the previous step.
```
