```python
import pandas as pd #Pandas is a library

#Word/Term                    Definition

#import pandas as pd           →   Loads the pandas library and gives
                                #it the shorthand alias "pd" for easier reference in code

#Pandas is a library          →   Inline comment noting that pandas is a library — used for data
                                #manipulation and analysis (reading CSVs, DataFrames, etc.)


#This cell imports pandas, the core library you'll use throughout the notebook for loading datasets and working with tabular data
#(DataFrames) — the starting point for most of your data science and ML projects.
```

```python
dataset=pd.read_csv("50_Startups.csv") #Dataset file name
dataset #Function call

#Word/Term                    Definition

#dataset = pd.read_csv         →   Loads the "50_Startups.csv" file
#("50_Startups.csv")               into a pandas DataFrame named dataset

#Dataset file name             →   Inline comment noting that "50_Startups.csv" is the name of
                                #the dataset file being loaded

#dataset                        →   Displays the contents of the dataset DataFrame — all rows and columns

#Function call                 →   Inline comment noting that this line simply calls/displays the
                                #dataset variable


#This cell loads the 50 Startups dataset (commonly used for multiple linear regression practice, with columns like R&D
#Spend, Administration, Marketing Spend, State, and Profit) and previews it to confirm the data loaded correctly before moving
#on to preprocessing and modeling.
```

```python
dataset=pd.get_dummies(dataset,drop_first=True) #Categorical data converted into numerical data. This technique is called One-Hot Encoding.
dataset

#Word/Term                    Definition

#dataset = pd.get_dummies      →   Converts categorical (text/object)
#(dataset, drop_first=True)        columns — like the 'State' column
                                #in this dataset — into numeric
                                #dummy/indicator variables,
                                #overwriting the original dataset variable

#drop_first=True                →   Removes the first category of
                                #each encoded column to avoid the
                                #dummy variable trap (multicollinearity
                                #caused by redundant columns)

#Categorical data              →   Inline comment correctly
#converted into numerical           explaining that this line
#data. This technique is           converts categorical data (like
#called One-Hot Encoding.          'State': New York, California,
                                #Florida) into numerical columns
                                #using One-Hot Encoding

#dataset                        →   Displays the updated dataset
                                #DataFrame, now with the 'State'
                                #column replaced by numeric dummy
                                #columns


#This cell converts the categorical 'State' column in the 50 Startups dataset into numeric dummy variables using One-Hot
#Encoding, dropping the first category to avoid multicollinearity — preparing the dataset for multiple linear regression, which
#requires all input features to be numeric.
```

```python
dataset.columns #Input dataset and output dataset - split

#Word/Term                    Definition

#dataset.columns               →   Displays the names of all columns
                                #in the dataset — useful for checking
                                #exact column names (including the
                                #new dummy columns created by
                                #get_dummies) before splitting into
                                #input and output

#Input dataset and             →   Inline comment noting the purpose
#output dataset - split            of this step — to identify column
                                #names so the dataset can next be
                                #split into independent features
                                #(input) and the target variable
                                #(output)


#This cell checks the exact column names after One-Hot Encoding was applied, so you can correctly select which columns become
#your independent features (X) and which one becomes your target variable (y — likely 'Profit') in the next step.
```

```python
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']] #Independent - Input dataset
dependent=dataset[["Profit"]] #Dependent - Output dataset

#Word/Term                    Definition

#independent = dataset         →   Selects 5 columns — 'R&D Spend',
#[['R&D Spend',                    'Administration', 'Marketing
#'Administration',                 Spend', 'State_Florida', and
#'Marketing Spend',                'State_New York' — as the
#'State_Florida',                  independent features (input)
#'State_New York']]                for the regression model

#Independent -                 →   Inline comment confirming this
#Input dataset                     variable holds the input/feature data

#dependent = dataset             →   Selects the 'Profit' column
#[["Profit"]]                       as the dependent/target variable
                                #(output) the model will predict

#Dependent -                    →   Inline comment confirming this
#The output dataset                     variable holds the output/target data


#This cell splits the dataset into independent features (R&D Spend, Administration, Marketing Spend, and the two
#State dummy columns) and the dependent target variable (Profit), preparing the data for multiple linear regression —
#this is the direct input/output split needed before train_test_split and model fitting.
```

```python
#Split - Training set and test set
from sklearn.model_selection import train_test_split  #Sklearn Supports Machine Learning
X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0)

#Word/Term                    Definition

#Split - Training set          →   Comment marking the start of the
#and test set                      train/test split section

#from sklearn.model_selection  →   Imports the function to split data
#import train_test_split          into training and testing subsets

#Sklearn Supports              →   Inline comment noting that
#Machine Learning                  scikit-learn (sklearn) is the
                                #library that supports machine
                                #learning tasks like this split

#X_train, X_test,              →   Splits independent (features) and
#y_train, y_test =                 dependent (target) data into
#train_test_split                  training and testing sets, with
#(independent, dependent,          30% of the data (test_size = 0.30)
#test_size=0.30,                   reserved for testing, and a fixed
#random_state=0)                   random_state for reproducible results


#This cell splits the 50 Startups dataset into training (70%) and testing (30%) sets, using the independent features (R&D Spend,
#Administration, Marketing Spend, State dummies) and the dependent variable (Profit) prepared earlier — setting up the data for
#training a multiple linear regression model.
```

```python
X_train

#Word/Term          Definition

#X_train       →     Displays the contents of X_train — the
                     #independent feature values (R&D Spend,
                     #Administration, Marketing Spend, State
                     #dummies) for the training set (70% of the data)


#This is a quick preview of X_train to visually confirm the training features look correct before fitting the regression model.
```

```python
#StandardScaler Preprocessing - Method
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)

#Word/Term                    Definition

#StandardScaler                →   Comment marking the start of the
#Preprocessing - Method            feature scaling section using StandardScaler

#from sklearn.preprocessing    →   Imports the StandardScaler class
#import StandardScaler            for standardizing features

#sc = StandardScaler()         →   Creates a StandardScaler object to
                                #standardize features (mean = 0, variance = 1)

#X_train = sc.fit_transform     →   Fits the scaler on training data
#(X_train)                          and transforms it (learns mean/std from training set only)

#X_test = sc.transform          →   Transforms test data using the
#(X_test)                           same scaler fitted on training data (prevents data leakage)


#This cell standardizes the independent features (R&D Spend, Administration, Marketing Spend, State dummies) before training
#the regression model — scaling ensures features with larger numeric ranges (like R&D Spend, often in the hundreds of
#thousands) don't disproportionately dominate the model compared to smaller-scale features like the State dummy variables (0/1).
```

```python
X_train

#Word/Term          Definition

#X_train       →     Displays the contents of X_train — the
                     #scaled independent feature values (after
                     #StandardScaler was applied) for the training set


#This is a quick preview of X_train to visually confirm the features were correctly standardized (should now show values
#centered around 0, rather than the original raw scale) before moving on to fitting the regression model.
```

```python
#Model creation
from sklearn.svm import SVR  #These first two lines are called as a procedure
regressor=SVR(kernel="linear") #Data behaviour - Pattern - Linear Pattern
regressor.fit(X_train,y_train) #Fit - Called Function. X-input, Y-output

#Word/Term                    Definition

#Model creation                →   Comment marking the start of the
                                #model creation section

#from sklearn.svm import SVR   →   Imports the Support Vector
                                #Regression class (SVR is the
                                #regression counterpart to SVC)

#These first two lines          →   Inline comment noting that the
#are called as a procedure         import and model creation lines
                                #together set up the "procedure" for
                                #building the regressor

#regressor = SVR                →   Creates an SVR model instance
#(kernel="linear")                  configured with a linear kernel,
                                #assuming a linear relationship
                                #between features and target

#Data behaviour -               →   Inline comment noting that the
#Pattern - Linear Pattern           'linear' kernel choice reflects
                                #an assumption that the data follows a linear pattern

#regressor.fit                  →   Trains the SVR model on the
#(X_train, y_train)                 scaled training features
                                #(X_train) and target values
                                #(y_train)

#Fit - Called Function.         →   Inline comment noting that .fit()
#X-input, Y-output                  is the function call that trains
                                #the model, taking X as input
                                #features and y as the output/target


#This cell creates and trains a Support Vector Regression (SVR) model with a linear kernel on the 50 Startups dataset — using
#the scaled R&D Spend, Administration, Marketing Spend, and State dummy variables to predict Profit, assuming the
#underlying relationship between these features and Profit is linear.
```

```python
regressor.intercept_

#Word/Term                    Definition

#regressor.intercept_          →   Retrieves the intercept (bias
                                #term / constant) of the trained
                                #SVR model — the value where the
                                #regression line/hyperplane crosses
                                #the y-axis when all features are 0


#This displays the intercept term learned by the SVR model, which is one part of the linear equation the model uses to
#predict Profit — useful alongside the coefficients (regressor.coef_, since the kernel is linear) to understand
#how each feature contributes to the prediction.
```

```python
regressor.n_support_

#Word/Term                    Definition

#regressor.n_support_          →   Retrieves the number of support
                                #vectors used by the trained SVR
                                #model — support vectors are the
                                #data points closest to (or within)
                                #the margin, which the model relies
                                #on most heavily to define the
                                #regression line/hyperplane


#This displays how many training samples were selected as support vectors by the SVR model — a smaller number relative to your
#total training set size (35 rows, since 30% of 50 was held out for testing) generally indicates a simpler, more generalizable
#model, while a very high number can suggest the model is struggling to fit the data with a clean margin.
```

```python
regressor.support_

#Word/Term                    Definition

#regressor.support_            →   Retrieves the indices (positions)
                                #of the training samples that were
                                #selected as support vectors by the
                                #trained SVR model — these correspond
                                #to specific rows in X_train that lie
                                #closest to (or within) the margin


#This displays which exact training samples (by index) act as support vectors for this SVR model — a useful follow-up to
#regressor.n_support_, since it shows you exactly which rows in your training data are driving the model's predictions, letting
#you trace those back to specific Startups in your original dataset if needed.
```

```python
y_pred=regressor.predict(X_test) #Evaluation Metric - For Model Evaluation

#Word/Term                    Definition

#y_pred = regressor.predict     →   Uses the trained SVR model to
#(X_test)                          predict Profit values for the
                                #scaled test features (X_test),
                                #storing the predicted values in y_pred

#Evaluation Metric -            →   Inline comment noting that this
#For Model Evaluation                predicted output will be used
                                #next to evaluate the model's
                                #performance (e.g. comparing y_pred
                                #against actual y_test values)


#This cell generates Profit predictions on the unseen test set using the trained SVR model — the next step will likely compare
#y_pred against the actual y_test values using metrics like R² score or Mean Squared Error to evaluate how well the linear-
#kernel SVR performs on this regression task.
```

```python
from sklearn.metrics import r2_score  #R2 Value Identification
r_score=r2_score(y_test,y_pred)

#Word/Term                    Definition

#from sklearn.metrics           →   Imports the r2_score function to
#import r2_score                   evaluate regression model performance

#R2 Value Identification        →   Inline comment noting the
                                #purpose of this import — to
                                #calculate the R² (R-squared) value

#r_score = r2_score              →   Calculates the R² score by
#(y_test, y_pred)                    comparing the actual Profit
                                #values (y_test) against the
                                #predicted Profit values (y_pred),
                                #storing the result in r_score


#This cell calculates the R² score for the linear-kernel SVR model — R² measures how much of the variance in Profit is
#explained by the model's predictions, with values closer to 1 (or 100%) indicating a better fit. This lets you quantify how
#well this SVR model performs on the 50 Startups dataset compared to other models or kernels you might try.
```

```python
r_score

#Word/Term          Definition

#r_score       →     Displays the R² (R-squared) score calculated
                     #in the previous cell — a value between 0 and
                     #1 (or negative, if the model performs worse
                     #than a horizontal mean line) showing how well
                     #the linear-kernel SVR model's predictions
                     #match the actual Profit values on the test set


#This is a quick check of r_score to see the model's actual R² value — the closer this is to 1 (e.g. 0.90+), the better the
#linear SVR model explains variation in Profit based on R&D Spend, Administration, Marketing Spend, and State.
```

```python
#User Input - Preprocessing Input
preinput=sc.transform([[1300,12000,4000,0,1]])
preinput

#Word/Term                Definition

#User Input -          →   Comment marking this section as
#Preprocessing Input     taking new user input and preparing
                          #it before prediction

#preinput =             →   Takes new raw input values (R&D
#sc.transform            Spend=1300, Administration=12000,
#([[1300,12000,           Marketing Spend=4000, State_Florida=0,
#4000,0,1]])              State_New York=1) and scales them
                          #using the SAME scaler (sc) fitted
                          #earlier on the training data

#preinput               →   Displays the scaled version of the
                          #input values


#Interview one-liner:
#"Before predicting on new data, I have to transform it using the same scaler that was fit on the training data — not a new 
#one — otherwise the values won't be on the same scale the model learned from, and predictions will be wrong."
```

```python
loaded_model=pickle.load(open("finalized_model_svm_regression.sav",'rb')) #Checking a saved model, whether working properly 
result=loaded_model.predict(preinput)

#Word/Term                Definition

#loaded_model =          →   Loads the previously saved SVR model
#pickle.load                 back from the file, in binary read
#(open("finalized_             mode ('rb')
#model_svm_regression.
#sav", 'rb'))

#Checking a saved       →   Comment noting the purpose — testing
#model, whether             whether the saved model actually
#working properly           works correctly after being reloaded

#result =                →   Uses the reloaded model to predict
#loaded_model.predict        Profit on the scaled input (preinput)
#(preinput)                  prepared earlier


#Summary:
#This cell reloads the saved SVR model from disk and tests it by predicting Profit on a new, already-scaled input. It confirms 
#that the model was saved and loaded correctly, and behaves the same way it would in a real deployment — where a model is 
#trained once, saved, and then reused for future predictions without retraining.
```
